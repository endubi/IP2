```python
### SQL IP
# We will first load the sql extension into our environment
%load_ext sql

# Then connect to our in memory sqlite database
# NB: This database will cease to exist as soon as the database connection is closed. 
%sql sqlite://
```

```python
# Importing Pandas and uploading Grand Electors data
import pandas as pd
with open('GrandElectors_by_state.csv','r') as f:
    GrandElect= pd.read_csv(f, index_col=0, encoding='utf-8')
%sql DROP TABLE if EXISTS GrandElect;
%sql PERSIST GrandElect;
%sql SELECT * FROM GrandElect LIMIT 5;
```

```python
## Uploading Population by state
with open('Population_by_state.csv','r') as f:
    Population = pd.read_csv(f, index_col=0, encoding='utf-8')
%sql DROP TABLE if EXISTS Population;
%sql PERSIST Population;
%sql SELECT * FROM Population LIMIT 5;
```

```python
# Converting Grand Electors State to Uppercase
%%sql
UPDATE GrandElect set State=UPPER(State);
SELECT * FROM GrandElect limit 5
```

```python
## Joining the two tables
%%sql
SELECT GrandElect.State, GrandElectors, Population
FROM GrandElect INNER JOIN  Population
ON GrandElect.State = Population.State;

```

```python
#### Creating a table from the joins
%%sql
CREATE TABLE Campaign
AS SELECT GrandElect.State, GrandElectors, Population
FROM GrandElect INNER JOIN  Population
ON GrandElect.State = Population.State;
```

```python
# Updating District of Columbia to DC
%%sql
UPDATE Campaign SET State="DC"
WHERE State="DISTRICT OF COLUMBIA"
```

```python
## Creating a new variable called ratio
%%sql
ALTER TABLE Campaign
ADD Ratio varchar; 
```

```python
### To compute the ratio between the number of grand electors and the population. 
%%sql
UPDATE Campaign SET Ratio =( CAST([GrandElectors] AS FLOAT) / [Population]);
SELECT * FROM Campaign LIMIT 5;
```

```python
# To order the states by decreasing ratio of Grand Electors per capita
%%sql
SELECT * FROM Campaign 
ORDER BY Ratio DESC;
```

```python
# Running total of Grand Electors
# To compute the half of the total of Grand Electors overall (in the whole country)
%%sql
SELECT SUM(GrandElectors)/2 
FROM Campaign
```
